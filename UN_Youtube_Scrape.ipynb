```python
# Loading all the necessary libraries for this project.
# Youtube is a DYNAMIC website. Which means that:
# To see more data, you will need to scroll down, which will take forever. 
# For this reason, in this project, we will use Selenium.
# Selenium will open a Chrome Browser and do the copy/paste thingy for us!

import sys, unittest, time, datetime
import urllib.request, urllib.error, urllib.parse
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import InvalidArgumentException
import pandas as pd
import csv
from selenium import webdriver
from selenium.common import exceptions
```

```python
# This can be applied to ANY Youtube channel, just change the link.
# When you run this line, a computer controlled Chrome will pop up.
# Let the computer do its thing. You can minimize and continue your work.
# Check the print if it's the desired output you want.
# IMPORTANT: You need to search "76th session" specifically with quotation marks...
# ...in order to find the UNGA 76th session videos from this mess.

url = 'https://www.youtube.com/c/unitednations/search?query=%2276th%20session%22'
channelid = url.split('/')[4]
driver = webdriver.Chrome()
driver.get(url)
time.sleep(15)
dt=datetime.datetime.now().strftime("%Y%m%d%H%M")
height = driver.execute_script("return document.documentElement.scrollHeight")
lastheight = 0

while True:
    if lastheight == height:
        break
    lastheight = height
    driver.execute_script("window.scrollTo(0, " + str(height) + ");")
    time.sleep(15)
    height = driver.execute_script("return document.documentElement.scrollHeight")

user_data = driver.find_elements_by_xpath('//*[@id="video-title"]')

video_list = []

for i in user_data:
    link = (i.get_attribute('href'))
    title = (i.text)
    print(title, link)
    vid_item = {
        'title': title,
        'link': link
    }
    
    video_list.append(vid_item)
    
driver.quit()
```

```python
# Let's create a DataFrame and check if our list looks okay.

df = pd.DataFrame(video_list)
```

```python
# If it's OK! Time to save it to CSV and complete the project.

df.to_csv('unga_76_youtube_data.csv')
```
